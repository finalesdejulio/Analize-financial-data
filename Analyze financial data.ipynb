```python
#1.Brief overview of company with those stocks code: Industry/sector (IT, pharmaceutical companies), # of Employees, market capital, main products, annual sales, book values, ...

#MSFT:
#Cng ty: Microsoft.
#Lnh vc: Cng ty cng ngh.
#Vn ha th trng: 2,43 nghn t USD
#Sn phm chnh: Microsoft Cloud; Microsoft Security; Dynamics 365; Microsoft 365; Microsoft Power Platform; ...
#Doanh thu hng nm: 2020: 49,36 t USD; 2021: 51,73 t USD; 2022: 198 t USD; qu 1 nm 2023: 52,9 t USD.
#Gi tr s sch: Gi tr s sch trn mi c phiu ca Microsoft trong qu 1 nm 2023 l 26,18 USD. Trong sut 13 nm qua; ch s P/B cao nht ca Microsoft l 16,95; thp nht l 3,09 v trung bnh l 7,70.

#TSM
#Cng ty: Taiwan Semiconductor Manufacturing Co.
#Lnh vc: chuyn v ch to cht bn dn.
#Vn ha th trng: 14,08 nghn t TWD.
#Sn phm chnh: Ultra Fast Recovery Rectifier (trr 15...30ns) [95] Schottky [37] ESD Protection Diode & Array [2] Super Junction [5] NPN Bipolar Transistor [1] PNP Bipolar Transistor [1] Standard Bridge Rectifier [467] Super Fast Recovery Bridge Rectifier (trr 35ns) [6] ...
#Doanh thu hng nm: 2020: 1,34 nghn t TWD; 2021: 1,59 nghn t TWD; 2022: 2.26 nghn t TWD; qu 1 nm 2023: 0,51 nghn t TWD
#Gi tr s sch: Gi tr s sch trn mi c phiu ca Taiwan Semiconductor Manufacturing Co trong qu 1 nm 2023 l 19,44 USD. Trong 13 nm qua; ch s P/B cao nht ca Taiwan Semiconductor Manufacturing Co l 9,21; thp nht l 2,91 v trung bnh l 3,80.

#GSK
#Cng ty: GlaxoSmithKline plc.
#Lnh vc: Cng ty dc phm.
#Vn ha th trng: 56,83 t GBP.
#Sn phm chnh: Thuc h st, gim au; thuc khng sinh; thuc iu tr hen suyn; thuc nga st rt;...
#Doanh thu hng nm: 2020: 34 t GBP; 2021: 34,1 t GBP; 2022: 29,3 t GBP; qu 1 nm 2023: 6,9 t GBP.
#Gi tr s sch: Gi tr s sch trn mi c phiu ca GSK trong qu 1 nm 2023 l 7,08 USD. Trong 13 nm qua; ch s P/B cao nht ca GSK l 491,88; thp nht l 3,26; v trung bnh l 9,28.




import math 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import datetime as dt
import plotly
import plotly.graph_objects as go
from matplotlib import style
from tqdm import tqdm


#2.Pull daily stock prices (open) from 01/Apr/2022 - 31/Mar/2023
df_stocks=['MSFT','TSM','GSK']
df_data=yf.download('MSFT,TSM,GSK',start='2022-4-1',end='2023-4-1')['Open']

print(df_data.head())
print(df_data.tail())
print(df_data.mean())
print(df_data.describe())



#3.Plot stocks prices (all 3) in one graph with all legend
style.use('ggplot')
plt.plot(df_data,label=['GSK','MSFT','TSM'])
plt.legend(loc='lower right')
plt.show()



#4.Plots histogram of returns of each stock (bin = 40)
df_MSFT = yf.Ticker('MSFT').history(period='1y',start='2022-4-1',end='2023-4-1').reset_index()[['Date','Open']]
df_MSFT_return=df_MSFT['Open'].pct_change(1)
style.use('ggplot')
plt.hist(df_MSFT_return,label='MSFT',bins=40)
plt.legend(loc='lower right')
plt.show()

df_TSM = yf.Ticker('TSM').history(period='1y',start='2022-4-1',end='2023-4-1').reset_index()[['Date','Open']]
df_TSM_return=df_TSM['Open'].pct_change(1)
style.use('ggplot')
plt.hist(df_TSM_return,label='TSM',bins=40)
plt.legend(loc='lower right')
plt.show()

df_GSK = yf.Ticker('GSK').history(period='1y',start='2022-4-1',end='2023-4-1').reset_index()[['Date','Open']]
df_GSK_return=df_GSK['Open'].pct_change(1)
style.use('ggplot')
plt.hist(df_GSK_return,label='GSK',bins=40)
plt.legend(loc='lower right')
plt.show()



#5.Plots Markowitzs bullet, Markowitzs Portfolio Frontier
df_return=df_data.pct_change().dropna()
df_return_mean=df_return.mean()*252
df_cov_matrix=df_return.cov()*252
df_num_portfolios=5000
df_results=np.zeros((3,df_num_portfolios))
for i in range(df_num_portfolios):
    df_weights=np.random.random(len(df_stocks))
    df_weights/=np.sum(df_weights)
    df_portfolio_return=np.sum(df_return_mean*df_weights)
    df_portfolio_std_dev=np.sqrt(np.dot(df_weights.T,np.dot(df_cov_matrix,df_weights)))
    df_results[0,i]=df_portfolio_return
    df_results[1,i]=df_portfolio_std_dev
    df_results[2,i]=df_portfolio_return/df_portfolio_std_dev
df_resultsdf=pd.DataFrame(df_results.T,columns=['Return','StdDev','Sharpe'])
plt.figure(figsize=(10,6))
plt.scatter(df_resultsdf['StdDev'],df_resultsdf['Return'],c=df_resultsdf['Sharpe'],cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Standard deviation')
plt.ylabel('Return')
plt.show()
```
