```python
#1.Brief overview of company with those stocks code: Industry/sector (IT, pharmaceutical companies), # of Employees, market capital, main products, annual sales, book values, ...

#MSFT:
#Công ty: Microsoft.
#Lĩnh vực: Công ty công nghệ.
#Vốn hóa thị trường: 2,43 nghìn tỉ USD
#Sản phẩm chính: Microsoft Cloud; Microsoft Security; Dynamics 365; Microsoft 365; Microsoft Power Platform; ...
#Doanh thu hằng năm: 2020: 49,36 tỉ USD; 2021: 51,73 tỉ USD; 2022: 198 tỉ USD; quý 1 năm 2023: 52,9 tỉ USD.
#Giá trị sổ sách: Giá trị sổ sách trên mỗi cổ phiếu của Microsoft trong quý 1 năm 2023 là 26,18 USD. Trong suốt 13 năm qua; chỉ số P/B cao nhất của Microsoft là 16,95; thấp nhất là 3,09 và trung bình là 7,70.

#TSM
#Công ty: Taiwan Semiconductor Manufacturing Co.
#Lĩnh vực: chuyên về chế tạo chất bán dẫn.
#Vốn hóa thị trường: 14,08 nghìn tỉ TWD.
#Sản phẩm chính: Ultra Fast Recovery Rectifier (trr 15...30ns) [95] Schottky [37] ESD Protection Diode & Array [2] Super Junction [5] NPN Bipolar Transistor [1] PNP Bipolar Transistor [1] Standard Bridge Rectifier [467] Super Fast Recovery Bridge Rectifier (trr 35ns) [6] ...
#Doanh thu hằng năm: 2020: 1,34 nghìn tỉ TWD; 2021: 1,59 nghìn tỉ TWD; 2022: 2.26 nghìn tỉ TWD; quý 1 năm 2023: 0,51 nghìn tỉ TWD
#Giá trị sổ sách: Giá trị sổ sách trên mỗi cổ phiếu của Taiwan Semiconductor Manufacturing Co trong quý 1 năm 2023 là 19,44 USD. Trong 13 năm qua; chỉ số P/B cao nhất của Taiwan Semiconductor Manufacturing Co là 9,21; thấp nhất là 2,91 và trung bình là 3,80.

#GSK
#Công ty: GlaxoSmithKline plc.
#Lĩnh vực: Công ty dược phẩm.
#Vốn hóa thị trường: 56,83 tỉ GBP.
#Sản phấm chính: Thuốc hạ sốt, giảm đau; thuốc kháng sinh; thuốc điều trị hen suyễn; thuốc ngừa sốt rét;...
#Doanh thu hằng năm: 2020: 34 tỉ GBP; 2021: 34,1 tỉ GBP; 2022: 29,3 tỉ GBP; quý 1 năm 2023: 6,9 tỉ GBP.
#Giá trị sổ sách: Giá trị sổ sách trên mỗi cổ phiếu của GSK trong quý 1 năm 2023 là 7,08 USD. Trong 13 năm qua; chỉ số P/B cao nhất của GSK là 491,88; thấp nhất là 3,26; và trung bình là 9,28.




import math 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import datetime as dt
import plotly
import plotly.graph_objects as go
from matplotlib import style
from tqdm import tqdm


#2.Pull daily stock prices (open) from 01/Apr/2022 - 31/Mar/2023
df_stocks=['MSFT','TSM','GSK']
df_data=yf.download('MSFT,TSM,GSK',start='2022-4-1',end='2023-4-1')['Open']

print(df_data.head())
print(df_data.tail())
print(df_data.mean())
print(df_data.describe())



#3.Plot stocks prices (all 3) in one graph with all legend
style.use('ggplot')
plt.plot(df_data,label=['GSK','MSFT','TSM'])
plt.legend(loc='lower right')
plt.show()



#4.Plots histogram of returns of each stock (bin = 40)
df_MSFT = yf.Ticker('MSFT').history(period='1y',start='2022-4-1',end='2023-4-1').reset_index()[['Date','Open']]
df_MSFT_return=df_MSFT['Open'].pct_change(1)
style.use('ggplot')
plt.hist(df_MSFT_return,label='MSFT',bins=40)
plt.legend(loc='lower right')
plt.show()

df_TSM = yf.Ticker('TSM').history(period='1y',start='2022-4-1',end='2023-4-1').reset_index()[['Date','Open']]
df_TSM_return=df_TSM['Open'].pct_change(1)
style.use('ggplot')
plt.hist(df_TSM_return,label='TSM',bins=40)
plt.legend(loc='lower right')
plt.show()

df_GSK = yf.Ticker('GSK').history(period='1y',start='2022-4-1',end='2023-4-1').reset_index()[['Date','Open']]
df_GSK_return=df_GSK['Open'].pct_change(1)
style.use('ggplot')
plt.hist(df_GSK_return,label='GSK',bins=40)
plt.legend(loc='lower right')
plt.show()



#5.Plots Markowitz’s bullet, Markowitz’s Portfolio Frontier
df_return=df_data.pct_change().dropna()
df_return_mean=df_return.mean()*252
df_cov_matrix=df_return.cov()*252
df_num_portfolios=5000
df_results=np.zeros((3,df_num_portfolios))
for i in range(df_num_portfolios):
    df_weights=np.random.random(len(df_stocks))
    df_weights/=np.sum(df_weights)
    df_portfolio_return=np.sum(df_return_mean*df_weights)
    df_portfolio_std_dev=np.sqrt(np.dot(df_weights.T,np.dot(df_cov_matrix,df_weights)))
    df_results[0,i]=df_portfolio_return
    df_results[1,i]=df_portfolio_std_dev
    df_results[2,i]=df_portfolio_return/df_portfolio_std_dev
df_resultsdf=pd.DataFrame(df_results.T,columns=['Return','StdDev','Sharpe'])
plt.figure(figsize=(10,6))
plt.scatter(df_resultsdf['StdDev'],df_resultsdf['Return'],c=df_resultsdf['Sharpe'],cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Standard deviation')
plt.ylabel('Return')
plt.show()
```
